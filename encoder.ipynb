```python
from sentence_transformers import CrossEncoder
import numpy as np

# Chargement d'un modèle Cross-Encoder entraîné spécifiquement pour le NLI
model = CrossEncoder('cross-encoder/nli-MiniLM2-L6-H768')
```

```python
# On passe les paires directement
paires = [
    ["Le chat est sur le tapis", "Le chat n'est pas sur le tapis"], #Contradiction
    ["L'accusé est coupable", "L'accusé n'est pas coupable"] #Contradiction
]

scores = model.predict(paires)

labels = ["Contradiction", "Entailment", "Neutral"]
for idx, sc in enumerate(scores):
    print(paires[idx])
    pred = labels[np.argmax(sc)]
    print(pred)
```
