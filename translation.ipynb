```python
import torch
import pandas as pd
from deep_translator import GoogleTranslator
```

```python
lang_mapping = {
    'english': 'en', 'french': 'fr', 'spanish': 'es', 'italien': 'it', 
    'german': 'de', 'portuguese': 'pt', 'dutch': 'nl', 'greek': 'el', 'russian': 'ru', 
    'japanese': 'ja', 'hindi': 'hi', 'chinese': 'zh-CN', 
    'egyptian': 'ar', 'thai': 'th'
}
```

```python
words = ["cat", "weather", "sock", "ergonomy", "tow", "reader", "modification", "big", "veil", "scratch", "desert",  "afternoon", "chocolate", "keyboard", "black", "boat", "tree", "chain", "lys", "virus", "apple", "ring", "beetroot", "electricity", "drum"]

translations = []

for w in words :
    t = []
    for lang_name, lang_code in lang_mapping.items():
        try:
            translated = GoogleTranslator(source='auto', target=lang_code).translate(w)
            t.append(translated.strip())
        except Exception:
            t.append(w) # Secours si plantage
    translations.append(t)


df = pd.DataFrame(translations, columns=lang_mapping.keys())
```
