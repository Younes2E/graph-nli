```python
from utils.document_spacyfier_younes import document_spacyfier, import_spacy_sentence_tokens_and_deps
from utils.text_span_v2 import document
```

```python

def dep(texte) :
    parser = document_spacyfier(lang='en')

    indexed_parsed_text = parser.process_document(texte)

    doc = document(nm='MonDocTest', content=texte)
    doc.constructions['sentences'] = {}

    curr_spacy_word_idx = 0
    while curr_spacy_word_idx < len(indexed_parsed_text):
        curr_spacy_word_idx, curr_sid = import_spacy_sentence_tokens_and_deps(
            i=curr_spacy_word_idx,
            spacy_word_indexed_obj_lst=indexed_parsed_text,
            doc=doc
        )



    print("--- EXAMEN DES MWU (Minimum Word Units) ---")
    for m_nm, m in doc.mwus.items():
        # m.annotations contient toutes les infos linguistiques
        mot = m.text(doc)
        pos = m.annotations['pos']
        dep = m.annotations['deprel']
        print(f"ID: {m_nm} | Mot: {mot:15} | POS: {pos:5} | Dep: {dep}")

    print("\n--- EXAMEN DES RELATIONS (Graphe de dépendance) ---")
    for r_nm, r in doc.rels.items():
        # r.src est le MWU gouverneur, r.trgt est le MWU dépendant, r.typ est le type de lien
        source = r.src.text(doc)
        cible = r.trgt.text(doc)
        print(f"Relation: {r.typ:10} | {source} ---> {cible}")
```

```python
dep("A dog was seen stealing food.")
```

```python
dep("John is eating pizza and pasta or Jane isn't playing football.")
```
